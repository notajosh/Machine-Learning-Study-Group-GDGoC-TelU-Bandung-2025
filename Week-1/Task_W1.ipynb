```python
# Make data.txt file and write with random numbers

file = open("data.txt", "w")
random_num = "42, 17, 8, 95, 33, 61, 29, 74, 12, 88, 5, 50, 23, 67, 91, 4, 39, 82, 56, 19, 77, 30, 48, 11, 64, 99, 2, 25, 70, 36"
file.write(str(random_num))
file.close()
```

```python
# Read data.txt file

with open("data.txt", "r") as file:
    print(file.read())
```

```python
# Split the comma, make the loop and make decision statement

list_num = random_num.split(", ")

for j in list_num:
  num = int(j)
  if num % 2 == 0:
    print(f"{num} adalah angka genap")
  else:
    print(f"{num} adalah angka ganjil")
```
