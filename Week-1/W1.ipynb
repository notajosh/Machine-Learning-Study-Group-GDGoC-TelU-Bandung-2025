```python
#Python Variable
nama = "budi"
umur = 21
print("Nama:", nama)
print(umur)

nilai_y = 15 #menggunakan underscore
print(nilai_y)
```

```python
#Tipe Data
data_integer = 11
print("data:", data_integer, "bertipe", type(data_integer))

data_float = 1.5
print("data:", data_float, "bertipe", type(data_float))

data_strings = "ucup"
print("data:", data_strings, "bertipe", type(data_strings))
```

```python
#List
mix = [1, "halo", True, 4,14]
print(mix)

buah = ["pisang", "alpukat", "jambu"]
print(len(buah))
buah.append("mangga")
print(buah)

print(len(buah))
```

```python
#Tuple
identitas = ("siti", 20, "jakarta")
print(identitas[0])
# identitas[0] = "budi" # Tidak bisa diakses (immutable)
```

```python
#Set
buah = {"pisang", "apel", "pisang", "jeruk"}
print(buah)
```

```python
#Dictionary
orang = {
    "nama": "budi",
    "umur": 20,
    "alamat": "jakarta"
}
print(orang["nama"])
orang["alamat"] = "bandung"
print(orang)
```

```python
#Casting
a = int(3.99)
b = float(3)
c = str(3)
print(a, b, c)
```

```python
x = 10
y = 5

print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x % y)
```

```python
x = 20
y = 2

print("x == y", x == y)
print("x != y", x != y)
print("x > y", x > y)
print("x < y", x < y)
print("x >= y", x >= y)
print("x <= y", x <= y)
```

```python
nilai = int(input("Masukkan Nilai Siswa:"))
if(90 <= nilai <= 100):
  print("Predikat A")
elif(80 <= nilai <= 89):
  print("Predikat B")
elif(65 <= nilai <= 79):
  print("Predikat C")
elif(40 <= nilai <= 64):
  print("Predikat D")
elif(0 <= nilai <= 39):
  print("Predikat E")
else:
  print("Nilai Tidak Valid"   )
```
